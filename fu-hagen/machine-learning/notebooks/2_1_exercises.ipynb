```python
# Calculate R-squared for a regression model
import numpy as np

# D={((1,12),5),((2,2),8),((2,9),16),((4,9),23)}
# θ=(−1,4,0.5)
D = [((1, 12), 5), ((2, 2), 8), ((2, 9), 16), ((4, 9), 23)]
X = np.array([list(xy) for xy, y in D])
y = np.array([y for xy, y in D])
theta = np.array([-1, 4, 0.5])
y_pred_fun = lambda x: theta[0] + theta[1] * x[0] + theta[2] * x[1]
y_pred = np.array([y_pred_fun(xy) for xy in X])
y_mean = np.mean(y)
ss_total = np.sum((y - y_mean) ** 2)
ss_residual = np.sum((y - y_pred) ** 2)
r_squared = 1 - (ss_residual / ss_total)
print(f"R-squared: {r_squared:.3f}")
```

```python
# Calculate quadratic loss for a regression model

# D={((12,0),−3),((19,1.5),−4),((40,2.9),−5),((40,4.3),−5)}
# θ=(−2,−0.05,−0.5)
D = [((12, 0), -3), ((19, 1.5), -4), ((40, 2.9), -5), ((40, 4.3), -5)]
X = np.array([list(xy) for xy, y in D])
y = np.array([y for xy, y in D])
theta = np.array([-2, -0.05, -0.5])
y_pred_fun = lambda x: theta[0] + theta[1] * x[0] + theta[2] * x[1]
y_pred = np.array([y_pred_fun(xy) for xy in X])
loss = np.sum((y - y_pred) ** 2)
print(f"Loss: {loss:.3f}")
```

```python
# Calculate loss with Tikhonov regularization for a regression model

# D={((182,23),172),((180,12),168),((162,26),153)}
# θ=(−15,1,0.4)
# λ=0.1

D = [((182, 23), 172), ((180, 12), 168), ((162, 26), 153)]
X = np.array([list(xy) for xy, y in D])
y = np.array([y for xy, y in D])
theta = np.array([-15, 1, 0.4])
lambda_ = 0.1
R_T = np.sum(theta[1:] ** 2)
print(f"Tikhonov regularization term: {R_T:.3f}")
y_pred_fun = lambda x: theta[0] + theta[1] * x[0] + theta[2] * x[1]
y_pred = np.array([y_pred_fun(xy) for xy in X])
loss = np.sum((y - y_pred) ** 2) + lambda_ * R_T
print(f"Loss with Tikhonov regularization: {loss:.3f}")
```

```python
# Find optimal model using scikit learn
# D={((23,36),0.1),((23,34),0),((26,29),1.2),((21,40),−0.5),((20,33),−1.5),((25,35),0.7)}
from sklearn.linear_model import LinearRegression

D = [((23, 36), 0.1), ((23, 34), 0), ((26, 29), 1.2), ((21, 40), -0.5), ((20, 33), -1.5), ((25, 35), 0.7)]
X = np.array([list(xy) for xy, y in D])
y = np.array([y for xy, y in D])
model = LinearRegression()
model.fit(X, y)
print("Optimal model coefficients:")
print(f"Intercept: {model.intercept_:.3f}")
print(f"Coefficients: {[round(coef, 3) for coef in model.coef_]}")
```

```python
# Find optimal model using scikit learn. Use polynomial features of degree 2
# D={((0.5,4.4),6.8),((1,4.0),8.1),((1.5,3.8),9.3),((1.6,3.7),11.1),((1.2,5.2),11.6),((0.7,4.3),13.6)}
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

D = [((0.5, 4.4), 6.8), ((1, 4.0), 8.1), ((1.5, 3.8), 9.3), ((1.6, 3.7), 11.1), ((1.2, 5.2), 11.6), ((0.7, 4.3), 13.6)]
D_poly = [((x[0], x[1], x[0] ** 2, x[0] * x[1], x[1] ** 2), y) for (x, y) in D]
X = np.array([list(xy) for xy, y in D_poly])
y = np.array([y for xy, y in D_poly])
model = LinearRegression()
model.fit(X, y)
print("Optimal model with polynomial features (degree 2) coefficients:")
theta = [round(float(x), 3) for x in np.concatenate(([model.intercept_], model.coef_))]
print(f"Coefficients: {theta}")
```

```python
# Find optimal model using scikit learn with ridge regression.
# D={((12,7),5.5),((10,8),6),((10,7.5),7),((15,5),6.5),((16,9),7.5),((18,8),10)}
# λ=0.1
from sklearn.linear_model import Ridge

D = [((12, 7), 5.5), ((10, 8), 6), ((10, 7.5), 7), ((15, 5), 6.5), ((16, 9), 7.5), ((18, 8), 10)]
X = np.array([list(xy) for xy, y in D])
y = np.array([y for xy, y in D])
model = Ridge(alpha=0.1)
model.fit(X, y)
print("Optimal model with ridge regression coefficients:")
theta = [round(float(x), 3) for x in np.concatenate(([model.intercept_], model.coef_))]
print(f"Coefficients: {theta}")
```

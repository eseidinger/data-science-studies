```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../datasets/sl_logreg_ex_rapid.csv")

plt.figure()
plt.scatter(df["age"], df["edible"])
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

X = df["age"].to_numpy().reshape(-1, 1)
y = df["edible"].to_numpy()

clf = LogisticRegression().fit(X,y)

xpts = np.linspace(0, 14, 100)

plt.figure()
plt.plot(xpts, clf.predict(xpts.reshape(-1,1)))
plt.scatter(df["age"], df["edible"])
plt.show()
```

```python
import sklearn.metrics as metrics

metrics.accuracy_score(y,clf.predict(X))
```

```python
metrics.precision_score(y,clf.predict(X))
```

```python
metrics.recall_score(y,clf.predict(X))
```

```python
df = pd.read_csv("../datasets/sl_logreg_ex_animals.csv")

X = df.iloc[:, 0:-1].values
y = df.iloc[:, 2].values

fig = plt.figure()

plt.scatter(df[y==0]["age"], df[y==0]["weight"], facecolor='red', marker='x')
plt.scatter(df[y==1]["age"], df[y==1]["weight"], facecolor='blue', marker='s')
plt.scatter(df[y==2]["age"], df[y==2]["weight"], facecolor='green', marker='+')

plt.show()
```

```python
from sklearn.inspection import DecisionBoundaryDisplay

#clf = LogisticRegression(multi_class='ovr').fit(X,y)
clf = LogisticRegression().fit(X,y)

plt.figure()

DecisionBoundaryDisplay.from_estimator(clf,X,plot_method="pcolormesh")

plt.scatter(df[y==0]["age"], df[y==0]["weight"], facecolor='red', marker='x')
plt.scatter(df[y==1]["age"], df[y==1]["weight"], facecolor='blue', marker='s')
plt.scatter(df[y==2]["age"], df[y==2]["weight"], facecolor='green', marker='+')

plt.show()
```

```python
metrics.accuracy_score(y,clf.predict(X))
```

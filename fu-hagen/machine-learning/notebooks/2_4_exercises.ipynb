```python
# Determine k nearest neighbor
# D={((−2,1),1),((−1.9,1.4),1),((−2.6,3.3),2),((−3.9,4.5),2),((−6.8,6),3)}
# x6=(−7.8,0.8)
 
import numpy as np

def knn(data, x, k):
    distances = []
    for point, label in data:
        distance = np.linalg.norm(np.array(point) - np.array(x))
        distances.append((distance, label))
    
    distances.sort(key=lambda x: x[0])
    neighbors = distances[:k]
    
    labels = [label for _, label in neighbors]
    return max(set(labels), key=labels.count)

# Given data points and their labels
data = [((-2, 1), 1),
        ((-1.9, 1.4), 1),
        ((-2.6, 3.3), 2),
        ((-3.9, 4.5), 2),
        ((-6.8, 6), 3)]
# New point to classify
x6 = (-7.8, 0.8)
# Number of nearest neighbors to consider
k = 1

# Determine the label for x6 using k-NN
predicted_label = knn(data, x6, k)
print(f"The predicted label for point {x6} is: {predicted_label}")
```

```python
# Determine k nearest neighbor regression
# D={((5,13),54),((7,29),80),((8,−48),48),((10,3),69),((11,−21),59)}
# x6=(9,59)
# k = 2

def knn_regression(data, x, k):
    distances = []
    for point, value in data:
        distance = np.linalg.norm(np.array(point) - np.array(x))
        distances.append((distance, value))
    
    distances.sort(key=lambda x: x[0])
    neighbors = distances[:k]
    
    values = [value for _, value in neighbors]
    return sum(values) / len(values)

# Given data points and their values
data_regression = [((5, 13), 54),
                   ((7, 29), 80),
                   ((8, -48), 48),
                   ((10, 3), 69),
                   ((11, -21), 59)]
# New point to predict
x6_regression = (9, 59)
# Number of nearest neighbors to consider
k_regression = 2
# Determine the predicted value for x6 using k-NN regression
predicted_value = knn_regression(data_regression, x6_regression, k_regression)
print(f"The predicted value for point {x6_regression} is: {predicted_value}")
```

```python
# Do a k nearest neighbor classification using scikit-learn
# D={((14.26,18.98),1),((9.45,8.61),1),((7.53,−0.7),1),((15.62,4.92),2),((17.47,−8.43),2),((6.66,−5.27),3)}
# x7=(13.92,−1.76)
# k = [1, 3, 5]

from sklearn.neighbors import KNeighborsClassifier

# Given data points and their labels
data_sklearn = [((14.26, 18.98), 1),
                ((9.45, 8.61), 1),
                ((7.53, -0.7), 1),
                ((15.62, 4.92), 2),
                ((17.47, -8.43), 2),
                ((6.66, -5.27), 3)]
# New point to classify
x7 = (13.92, -1.76)
# Prepare the data for scikit-learn
X = np.array([point for point, label in data_sklearn])
y = np.array([label for point, label in data_sklearn])
# Create and fit the KNN classifier

for k in [1, 3, 5]:
    knn_classifier = KNeighborsClassifier(n_neighbors=k)
    knn_classifier.fit(X, y)
    predicted_label_sklearn = knn_classifier.predict([x7])
    print(f"The predicted label for point {x7} with k={k} is: {predicted_label_sklearn[0]}")
```

```python
# Calculate logistic loss function
# D={((56,44),1),((50,37),0),((47,53),1),((18,40),0)}
# θ=(45,−0.5,−0.5)

import math
def logistic_loss(theta, data):
    total_loss = 0
    for x, y in data:
        z = theta[0] + theta[1] * x[0] + theta[2] * x[1]
        p = 1 / (1 + math.exp(-z))
        loss = -y * math.log(p) - (1 - y) * math.log(1 - p)
        total_loss += loss
    return total_loss

# Given data and parameters
data = [((56, 44), 1), ((50, 37), 0), ((47, 53), 1), ((18, 40), 0)]
theta = (45, -0.5, -0.5)
loss = logistic_loss(theta, data)
print(f"Logistic loss: {loss:.3f}")
```

```python
# Determine the logistic model using scikit-learn
# D={((16,22),1),((43,70),1),((135,50),0),((42,46),0),((60,59),0),((105,140),1)}

from sklearn.linear_model import LogisticRegression
import numpy as np
# Prepare the data
data = [((16, 22), 1), ((43, 70), 1), ((135, 50), 0), ((42, 46), 0), ((60, 59), 0), ((105, 140), 1)]
X = np.array([x for x, y in data])
y = np.array([y for x, y in data])
# Fit the logistic regression model
model = LogisticRegression()
model.fit(X, y)
# Print the coefficients
theta = [round(float(x), 3) for x in np.concatenate((model.intercept_, model.coef_[0]))]
print(f"Logistic model parameters (theta): {theta}")
```

```python
# Determine the logistic model using scikit-learn and polynomial features of degree 2
# D={((12,11),0),((13,14),1),((11,8),1),((6,2),0),((12,10),0),((7,11),0)}

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LogisticRegression
import numpy as np
# Prepare the data
data = [((12, 11), 0), ((13, 14), 1), ((11, 8), 1), ((6, 2), 0), ((12, 10), 0), ((7, 11), 0)]
data_poly = [((x[0], x[1], x[0]**2, x[0]*x[1], x[1]**2), y) for (x, y) in data]
X = np.array([x for x, y in data_poly])
y = np.array([y for x, y in data_poly])
# Create polynomial features
model = LogisticRegression()
model.fit(X, y)
# Print the coefficients rounded to 3 decimal places
theta = [round(float(x), 3) for x in np.concatenate((model.intercept_, model.coef_[0]))]
print(f"Logistic model parameters (theta): {theta}")
```

```python
# Determine multiclass logistic regression model using scikit-learn
# D={((1,80),2),((20,3),2),((5,0),1),((13,12),1),((60,30),3),((10,40),3)}
from sklearn.linear_model import LogisticRegression
import numpy as np
# Prepare the data
data = [[(1, 80), 2], [(20, 3), 2], [(5, 0), 1], [(13, 12), 1], [(60, 30), 3], [(10, 40), 3]]

# set all y values to 0 where y is not 1, and to 1 where y is 1
D_ = []
D_.append([[x, 1 if y == 1 else 0] for x, y in data])
# set all y values to 0 where y is not 2, and to 1 where y is 2
D_.append([[x, 1 if y == 2 else 0] for x, y in data])
# set all y values to 0 where y is not 3, and to 1 where y is 3
D_.append([[x, 1 if y == 3 else 0] for x, y in data])

for i, d_ in enumerate(D_):
    print(f"Training model for class {i+1}")
    print(f"Data: {d_}")
    X = np.array([x for x, y in d_])
    y = np.array([y for x, y in d_])
    model = LogisticRegression()
    model.fit(X, y)
    theta = [round(float(x), 3) for x in np.concatenate((model.intercept_, model.coef_[0]))]
    print(f"Logistic model parameters (theta): {theta}")
```

```python
# Determine the logistic model using scikit-learn and polynomial features of degree 2
# D={((0.3,0.3),1),((0.24,0.26),4),((0.33,0.2),2),((0.35,0.4),3),((0.39,0.38),3),((0.42,0.26),1)}

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LogisticRegression
import numpy as np

# Prepare the data
data = [[(0.3, 0.3), 1], [(0.24, 0.26), 4], [(0.33, 0.2), 2], [(0.35, 0.4), 3], [(0.39, 0.38), 3], [(0.42, 0.26), 1]]
data_poly = [[(round(x[0], 3), round(x[1], 3), round(x[0]**2, 3), round(x[0]*x[1], 3), round(x[1]**2, 3)), y] for x, y in data]
print(f"Polynomial features: {data_poly}")

for i in range(1, 5):
    print(f"Training model for class {i}")
    # set all y values to 0 where y is not i, and to 1 where y is i
    d_ = [[x, 1 if y == i else 0] for x, y in data_poly]
    print(f"Data: {d_}")
    X = np.array([x for x, y in d_])
    y = np.array([y for x, y in d_])
    # Fit the logistic regression model
    model = LogisticRegression()
    model.fit(X, y)
    theta = [round(float(x), 3) for x in np.concatenate((model.intercept_, model.coef_[0]))]
    print(f"Model parameters rounded to 3 decimals (theta): {theta}")
```

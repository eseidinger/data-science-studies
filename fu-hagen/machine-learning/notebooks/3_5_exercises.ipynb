```python
# E={(−1.222,−1.297,−1.648),(−0.604,−0.208,1.436),(1.247,0.809,−0.414),(1.247,1.826,−0.846),(−0.693,0.301,1.128),(−0.869,−1.006,0.141),(0.894,−0.425,0.203)}
# Perform a principal component analysis (PCA) on the data points in E and reduce the dimensionality to 2. Give the eigenvalues.

import numpy as np
from sklearn.decomposition import PCA

# Define the data points
E = np.array([[-1.222, -1.297, -1.648], [-0.604, -0.208, 1.436], [1.247, 0.809, -0.414], [1.247, 1.826, -0.846], [-0.693, 0.301, 1.128], [-0.869, -1.006, 0.141], [0.894, -0.425, 0.203]])
# Perform PCA and reduce dimensionality to 2
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(E)
# Get the eigenvalues
eigenvalues = pca.explained_variance_
print("Reduced Data:\n", reduced_data)
print("Eigenvalues:\n", eigenvalues)
```

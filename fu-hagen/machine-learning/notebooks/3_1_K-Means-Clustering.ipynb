```python
import matplotlib.pyplot as plt
import pandas as pd

X = pd.read_csv("../datasets/dataset_movies2.csv").values
X
```

```python
plt.figure()
plt.scatter(X.T[0], X.T[1])
plt.show()
```

```python
from sklearn.cluster import KMeans

clust = KMeans(n_clusters=3,n_init=10).fit(X)
y = clust.predict(X)

plt.figure()

plt.scatter(X[y==0].T[0], X[y==0].T[1])
plt.scatter(X[y==1].T[0], X[y==1].T[1])
plt.scatter(X[y==2].T[0], X[y==2].T[1])

plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

X_scaled = StandardScaler().fit_transform(X)

clust = KMeans(n_clusters=3,n_init=10).fit(X_scaled)
y = clust.predict(X_scaled)

plt.figure()

plt.scatter(X[y==0].T[0], X[y==0].T[1])
plt.scatter(X[y==1].T[0], X[y==1].T[1])
plt.scatter(X[y==2].T[0], X[y==2].T[1])

plt.show()
```

```python
# Use scikit-learn to do a naive bayes classification
# D={((0,50),0),((1,200),2),((2,200),1),((0,100),1),((1,50),0),((0,200),2)}
# x7=(2,50)

from sklearn.naive_bayes import GaussianNB
import numpy as np

# Define the dataset
X = np.array([[0, 50],
              [1, 200],
              [2, 200],
              [0, 100],
              [1, 50],
              [0, 200]])
y = np.array([0, 2, 1, 1, 0, 2])
# Create a Gaussian Naive Bayes classifier
clf = GaussianNB()
# Fit the classifier to the data
clf.fit(X, y)
# Make a prediction
x7 = np.array([[2, 50]])
prediction = clf.predict(x7)
print(f'The predicted class for {x7[0]} is: {prediction[0]}')
```

```python
# Use scikit-learn to do a naive bayes classification
# D_train={((1.29,−1.83),1),((0.76,−0.95),1),((5.24,−1.59),0),((8.74,−0.27),0),((2.77,−2.23),0),((1.24,−1.65),1)}
# D_test={((0.62,−1.58),1),((6.82,0.07),0),((2.64,−0.15),1),((3.23,−1.58),0)}
# evaluate the accuracy, precision, recall, f1-score of the classifier on D_test

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Define the training dataset
X_train = np.array([[1.29, -1.83],
                    [0.76, -0.95],
                    [5.24, -1.59],
                    [8.74, -0.27],
                    [2.77, -2.23],
                    [1.24, -1.65]])
y_train = np.array([1, 1, 0, 0, 0, 1])
# Define the test dataset
X_test = np.array([[0.62, -1.58],
                     [6.82, 0.07],
                     [2.64, -0.15],
                     [3.23, -1.58]])
y_test = np.array([1, 0, 1, 0])
# Create a Gaussian Naive Bayes classifier
clf2 = GaussianNB()
# Fit the classifier to the training data
clf2.fit(X_train, y_train)
# Make predictions on the test data
y_pred = clf2.predict(X_test)
print(f'Predictions on test set: {y_pred}')
# Evaluate the classifier
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
# Print the evaluation metrics rounded to 3 decimal places
print(f'Accuracy: {accuracy:.3f}')
print(f'Precision: {precision:.3f}')
print(f'Recall: {recall:.3f}')
print(f'F1-score: {f1:.3f}')
```

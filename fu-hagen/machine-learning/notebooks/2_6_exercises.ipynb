```python
# Use scikit-learn to do a decision tree classification
# D_train={((37.5,2),1),((36,1),0),((39,1),1),((37.5,1),0),((39,0),1),((37.5,0),0),((37.5,1),0),((36,2),1)}
# D_test={((36,0),0),((37.5,1),0),((39,1),1),((39,2),0)}

from sklearn import tree
import numpy as np
from sklearn.metrics import confusion_matrix

# Training data
X_train = np.array([[37.5, 2], [36, 1], [39, 1], [37.5, 1], [39, 0], [37.5, 0], [37.5, 1], [36, 2]])
y_train = np.array([1, 0, 1, 0, 1, 0, 0, 1])
# Test data
X_test = np.array([[36, 0], [37.5, 1], [39, 1], [39, 2]])
y_test = np.array([0, 0, 1, 0])
# Create a decision tree classifier
clf = tree.DecisionTreeClassifier()
# Fit the model to the training data
clf = clf.fit(X_train, y_train)
# Predict the labels for the test data
y_pred = clf.predict(X_test)
# Print the predicted labels
print("Predicted labels:", y_pred)
# Print the actual labelsprint("Actual labels:", y_test)
print("Actual labels:", y_test)
# Create a confusion matrixfrom sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
```

```python
# Use scikit-learn to do a decision tree classification
# use a decision forest classifier instead of a decision tree classifier
# D_train={((37.5,2),1),((36,1),0),((39,1),1),((37.5,1),0),((39,0),1),((37.5,0),0),((37.5,1),0),((36,2),1)}
# D_test={((36,0),0),((37.5,1),0),((39,1),1),((39,2),0)}
# Calculate accuracy and f1-score

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score

n_estimators = [100, 10, 1000]
for n in n_estimators:
# Create a random forest classifier
    rf_clf = RandomForestClassifier(n_estimators=n, random_state=42)
    # Fit the model to the training data
    rf_clf.fit(X_train, y_train)
    # Predict the labels for the test data
    y_rf_pred = rf_clf.predict(X_test)
    # Print the predicted labels
    print("Random Forest Predicted labels:", y_rf_pred)
    # Print the actual labels
    print("Actual labels:", y_test)
    # Calculate accuracy and f1-score
    accuracy = accuracy_score(y_test, y_rf_pred)
    f1 = f1_score(y_test, y_rf_pred)
    print("Random Forest Accuracy:", accuracy)
    print("Random Forest F1 Score:", f1)
```

```python
# Platz für Tests und Analysen
for col in df_mundraub:
    cValues= df_mundraub[col].dropna()
    if cValues.count()-cValues.nunique() > 0:
        print("\n>>",col, "values:", cValues.count(), "\n",
              "Unique values:", cValues.nunique(), "\n",
              "Duplicated values:", cValues.count() - cValues.nunique(), "\n",
              "Prozent duplicated:",100*(cValues.count() - cValues.nunique())/cValues.count())
    else:
        print("\n>>",col, "is complete and unique")
```

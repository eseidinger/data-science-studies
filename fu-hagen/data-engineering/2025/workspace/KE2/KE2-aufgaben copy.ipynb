## Refill-Deutschland's Merkmale Analyze
Bestimmen Sie, wenn möglich, die statistischen Merkmale der numerischen Spalten. Tragen Sie die erhaltenen Informationen in die nachstehende Tabelle ein.

*Hinweis:* Wenn Sie weitere Tipps benötigen, verwenden Sie den Schieberegler unten.

- Unter der Kategorie _**Möglicher Datentyp**_: tragen Sie die primitiven Datentypen (Text, Number, Boolean, Date, usw.) ein.
- Für die Berechnung der _**Anzahl Fehlende Werte**_ können Sie die folgende Formel verwenden:
$Fehlende Werte = Gesamtzahl der Werte − Verfügbare Werte$
- Für die Berechnung der _**Anzahl von wiederholten Werten**_ können Sie die folgende Formel verwenden:
$Wiederholte Werte = Verfügbare Werte − Anzahl einzigartiger Werte$

```python
# Platz für Tests und Analysen

def calculate_non_unique_percentage(column):
    total = df_refill[column].size
    non_unique = total - df_refill[column].nunique()
    return non_unique, non_unique / total * 100 if total > 0 else 0

print(calculate_non_unique_percentage("title"))

def count_missing_values(column):
    return df_refill[column].isnull().sum()

print(count_missing_values("email"))
```

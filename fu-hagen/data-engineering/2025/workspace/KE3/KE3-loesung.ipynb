Der erste reguläre Ausdruck, `regExp1` wird verwendet, um unerwünschte Zeichen wie Klammern oder Trennzeichen aus Telefonnummern zu entfernen. Die Ausdrücke `regExp2` und `regExp3` nutzen den Anker "^", um sicherzustellen, dass die Nummer mit "+49" beginnt.

Verwenden Sie die regulären Ausdrücke `regExp1`, `regExp2` und `regExp3`, um die jeweiligen Werte zu bereinigen und alle Telefonnummern in ein einheitliches Format zu überführen.

# Fehlende Werte
Oft fehlen Werte in den Daten. Dies kann verschiedene Gründe haben, z.B. kann ein Wert absichtlich weggelassen werden (z.B. der Titel in einer Anrede) oder die Daten können nicht korrekt übertragen werden. Für viele Auswertungen ist es jedoch wichtig, dass es keine fehlenden Werte gibt.

Das Ersetzen fehlender Werte wird als (Missing Value) Imputation bezeichnet. Hierfür  gibt es verschiedene Strategien: 

- **_Wert ermitteln_**: Im besten Fall kann der korrekte Wert ermittelt werden, bspw. durch Rücksprache mit (Fach-)Expert*innen. Dies ist jedoch nur selten der Fall und häufig mit großem Aufwand verbunden. 
- **_Zeilen löschen_**: Wenn nur wenige Einzelwerte fehlen, kann es sinnvoll sein, die entsprechenden Zeilen zu löschen. Hierbei muss jedoch darauf geachtet werden, dass nicht zu viele Informationen verloren gehen. 
- **_Spalte löschen_**: Wenn in einer Spalte so viele Werte fehlen, dass sie diese keine Aussagekraft mehr hat, kann es sinnvoll sein, die gesamte Spalte zu entfernen. 
- **_Neuen Wert_**: Fehlen nur wenige Einzelwerte, so kann ein neuer Wert eingefügt werden.
- **_Neue Kategorie_**: Bei nominalen Werten kann eine neue Kategorie (z.B. ”Unbekannt”) hinzugefügt werden. 
- **_Mode Imputation_**: Fehlende Werte werden durch den am häufigsten Wert ersetzt. 
- **_Mean Imputation_**: Fehlende Werte werden durch den Mittelwert ersetzt. 
- **_Median Imputation_**: Fehlende Werte werden durch den Median ersetzt. 
Mode, Mean und Median Imputation können auch auf Basis der Werte einer anderen Spalte durchgeführt werden.
- **_Imputation mit_** 0 oder einem anderen **_konstanten Wert_**: Fehlende Werte werden durch eine Konstante ersetzt, z.B durch 0 oder eine andere Konstante, je nach Anwendung.
- **_Modellbasierte Imputation_**: Ein maschinelles Lernmodell wird darauf trainiert, die fehlenden Werte vorherzusagen. 

Zusätzlich zu den oben genannten Strategien ist es möglich, **_ein neues Feature_** zu erstellen, das anzeigt, ob eine Imputation durchgeführt wurde (=1) oder nicht (=0).

## Déjà vu :: Duplikate

Eine weitere Spalte, in der viele Werte fehlen, ist die Spalte `openinghours`. Im Gegensatz zu den vorhergehenden Spalten können die Werte aus der Spalte `description` entnommen werden.

*Hinweis: Diese Aufgabe kann aufgrund ihrer Komplexität in drei Niveaustufen bearbeitet werden (Anfängerniveau, Durchschnittsniveau, Expertenniveau).*

*_1. Anfängerniveau_: Auf diesem Niveau können Sie mithilfe von Pseudocode beschreiben, wie die Aufgabe gelöst werden könnte.*

*_2. Durchschnittsniveau_: Auf diesem Niveau kann man mit verschachtelten ifs und vielen Zeilen Code gelöst werden.*

*_3. Expertenniveau_: Auf dieser Niveau können Sie reguläre Ausdrücke verwenden. Wenn Sie diese Niveau auswählen, können Sie den Schieberegler unten verwenden, um den regulären Ausdruck anzuzeigen. Anschließend , können Sie die Codezelle vervollständigen, um die in dieser Aufgabe vorgesehenen Änderungen vorzunehmen.*

*Die Aufgabe muss nur mit einer Variante gelöst werden. Die zweite Variante (Durchschnittsniveau) ist dabei die umständlichste, weswegen wir aus Zeitgründen eher Variante 1 oder 3 empfehlen.*

```python
# Regular expression to find “Öffnungszeiten” or derivatives (can be improved)
import re

# Keywords
OPENING_TAGS = re.compile(r'\b(Öffnungszeiten:?|Wir haben für Sie geöffnet):?\b', re.IGNORECASE)

# Keywords for Weekdays and 'FT' Feiertagen
DAYS = r'(?:Mo|Di|Mi|Do|Fr|Sa|So|FT)'

# Valid openinghours-blocks
# Days can be individual, in range separated by '-' and not consecutive separated by ';'.
OPENING_PATTERN = re.compile(rf'''
    (?:
    
        {DAYS}(?:-{DAYS})?(?:;{DAYS})* \s+           # Days         
    )?
    \d{{1,2}}[:.]\d{{2}}                             # Start time
    (?:-\d{{1,2}}[:.]\d{{2}})?                       # End time
    (?:;\d{{1,2}}[:.]\d{{2}}-\d{{1,2}}[:.]\d{{2}})*  # additional times separated by ';'
''', re.VERBOSE)
```

```python
# Ihre Lösung :: Für Expertenniveau

def extract_openingHours(text, openingH):
    if not isinstance(text, str):  
        text = ""

    # Remove tags like 'Öffnungszeiten'
    clean_text = OPENING_TAGS.sub('', text).strip()

    # Find time blocks
    opening_matches = OPENING_PATTERN.findall(clean_text)
    if opening_matches:
    
        opening_text = ' '.join(opening_matches).strip()

        # Save description
        description_text = OPENING_PATTERN.sub('', clean_text).strip()
        
        description_text = re.sub(r'^[\W_]+$', '', description_text.strip())
        return description_text, opening_text
    else:
        return text, openingH

df_refill[['description', 'openinghours']] = df_refill.apply(
    lambda row: pd.Series(extract_openingHours(row.description, row.openinghours)),
    axis=1)
```

## Paket unzustellbar II :: Zusatzaufgabe
Bestimmen Sie die fehlenden Werte. Diese Informationen können mit Hilfe der vorhandenen geographischen Punkte extrahiert werden, siehe Spalte `geom`.

*Hinweis: Dies ist eine freiwillige Aufgabe. Falls Sie diese bearbeiten möchten, nutzen Sie den Schieberegler unten für weitere Hinweise*

<a href="https://geopy.readthedocs.io/en/stable/">GeoPy</a> ist eine Python-Bibliothek zur Lokalisierung von Adressen, Städten, Ländern und Landmarken auf der ganzen Welt basierend auf Geokodern von Drittanbietern und anderen Datenquellen. Zu den Geokodierern gehören OpenStreetMap Nominatim, Google Geocoding API (V3) und andere. 

Bitte beachten Sie, dass das GeoPy-Modul über das Internet auf Dritte angewiesen ist. Die Beschaffung der fehlenden Daten kann daher einige Zeit in Anspruch nehmen. Außerdem kann es je nach Internetverbindung dazu kommen, dass einzelne Datenpunkte nicht abgefragt werden können, entsprechende Fehlermeldungen können Sie ignorieren. Um das Modul zu installieren, führen Sie bitte die folgende Zelle aus:

```python
# Ihre Lösung
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut

# Geokodierungsdienst einrichten
geolocator = Nominatim(user_agent="mundraub")

def getAddressPlzCity(geoLoc):
    """
    Extract the city, postal code, and street from a geographical location.
    Implements caching to avoid repeated geocoding requests.
    """
    try:
        # POINT (8.263338 48.993153) -> get longitude and latitude
        longitude, latitude = geoLoc[7:-1].split(" ")
        location = geolocator.reverse((latitude, longitude), 
                                      exactly_one=True)
        val = location.raw.get('address', {})
        
        city = val.get('city') or val.get('town') or val.get('municipality') or val.get('village')
        postal_code = val.get('postcode')
        street = val.get('road')
        
        if postal_code is not None:
            postal_code = float(postal_code)
        
        return city, postal_code, street
    
    except GeocoderTimedOut:
        # Handle geocoding timeouts by returning None values
        return None, None, None
    except Exception as e:
        # Handle any other exceptions and return None values
        print(f"Error processing {geoLoc}: {e}")
        return None, None, None
    
def setMissingValuesInDS(dSet):
    """
    set missing values in the given dataset
    """
    for index, row in dSet.iterrows():
        # row with missing value?
        need_geocode = pd.isna(row['city']) or pd.isna(row['plz']) or pd.isna(row['street'])
        
        if need_geocode:
            nCi, nPl, nSt = getAddressPlzCity(row.geom)
            
        if pd.isna(row['city']):
            df_mundraub.at[index, 'city'] = nCi
        if pd.isna(row['plz']):
            df_mundraub.at[index, 'plz'] = nPl
        if pd.isna(row['street']):
            df_mundraub.at[index, 'street'] = nSt
            
# missing: Zip-code(476) City(473) Street(896)
setMissingValuesInDS(df_mundraub)
```

```python
# Ihre Lösung
from datetime import datetime

# Extracts the date and time components from a given string.
def extract_datetime(date_str):
    # Find the position of the first '-' character
    pos = date_str.find('-')
    
    date_part = date_str[4:pos-1]  # Extract date 'dd.mm.yyyy'
    time_part = date_str[pos+2:pos+7]  # Extract time 'HH:MM'
    
    return f'{date_part} {time_part}'

# Clean 'created_at' column
df_mundraub['created_at'] = df_mundraub['created_at'].map(extract_datetime)

# Set the right datatype
df_mundraub['created_at'] = pd.to_datetime(df_mundraub['created_at'], 
                                           format="%d.%m.%Y %H:%M")
```

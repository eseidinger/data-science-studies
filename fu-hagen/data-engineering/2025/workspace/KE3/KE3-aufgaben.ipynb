# Fehlende Werte
Oft fehlen Werte in den Daten. Dies kann verschiedene Gründe haben, z.B. kann ein Wert absichtlich weggelassen werden (z.B. der Titel in einer Anrede) oder die Daten können nicht korrekt übertragen werden. Für viele Auswertungen ist es jedoch wichtig, dass es keine fehlenden Werte gibt.

Das Ersetzen fehlender Werte wird als (Missing Value) Imputation bezeichnet. Hierfür  gibt es verschiedene Strategien: 

- **_Wert ermitteln_**: Im besten Fall kann der korrekte Wert ermittelt werden, bspw. durch Rücksprache mit (Fach-)Expert*innen. Dies ist jedoch nur selten der Fall und häufig mit großem Aufwand verbunden. 
- **_Zeilen löschen_**: Wenn nur wenige Einzelwerte fehlen, kann es sinnvoll sein, die entsprechenden Zeilen zu löschen. Hierbei muss jedoch darauf geachtet werden, dass nicht zu viele Informationen verloren gehen. 
- **_Spalte löschen_**: Wenn in einer Spalte so viele Werte fehlen, dass sie diese keine Aussagekraft mehr hat, kann es sinnvoll sein, die gesamte Spalte zu entfernen. 
- **_Neuen Wert_**: Fehlen nur wenige Einzelwerte, so kann ein neuer Wert eingefügt werden.
- **_Neue Kategorie_**: Bei nominalen Werten kann eine neue Kategorie (z.B. ”Unbekannt”) hinzugefügt werden. 
- **_Mode Imputation_**: Fehlende Werte werden durch den am häufigsten Wert ersetzt. 
- **_Mean Imputation_**: Fehlende Werte werden durch den Mittelwert ersetzt. 
- **_Median Imputation_**: Fehlende Werte werden durch den Median ersetzt. 
Mode, Mean und Median Imputation können auch auf Basis der Werte einer anderen Spalte durchgeführt werden.
- **_Imputation mit_** 0 oder einem anderen **_konstanten Wert_**: Fehlende Werte werden durch eine Konstante ersetzt, z.B durch 0 oder eine andere Konstante, je nach Anwendung.
- **_Modellbasierte Imputation_**: Ein maschinelles Lernmodell wird darauf trainiert, die fehlenden Werte vorherzusagen. 

Zusätzlich zu den oben genannten Strategien ist es möglich, **_ein neues Feature_** zu erstellen, das anzeigt, ob eine Imputation durchgeführt wurde (=1) oder nicht (=0).

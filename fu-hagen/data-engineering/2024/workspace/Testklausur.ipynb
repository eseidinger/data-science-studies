## Pandas

```python
data = [["Luigis Rennpiste", "Luigi"], ["Kuhmuh-Farm", "Prinzessin Peach"], ["Kuupa-Strand", "Donkey Kong"], ["Kalimari-Wüste", "Luigi"]]
columns = ["Streckenname", "Besitzer:in"]
mario_kart_64_pilz_cup = pd.DataFrame(data=data, columns=columns)
```

## Visualisierung
**Achtung!** Sie benötigen nur EINE der beiden Bibliotheken, Sie können selbst wählen, ob Sie die Aufgaben mit `matplotlib` oder mit `seaborn` lösen wollen.

### Matlplotlib

```python
mario_kart_64_pilz_cup \
    .groupby("Besitzer:in") \
    .size() \
    .sort_values() \
    .plot.bar(figsize=(10,5), title="Anzahl Strecken pro Besitzer:in", xlabel="Besitzer:in", ylabel="Anzahl Strecken")
plt.show()
```

### Seaborn

```python
sns.countplot(x="Besitzer:in", data=mario_kart_64_pilz_cup)
plt.xlabel("Besitzer:in")
plt.ylabel("Anzahl Strecken")
plt.title("Anzahl Strecken pro Besitzer:in")
plt.show()
```
